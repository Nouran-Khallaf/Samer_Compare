### 3.Compring Data

### 3.1 Creat dictionary of the common words with both classification

```python
BUC = Buckwalter_dataset.to_dict('index')
CER = CEFR_dataset.to_dict('index')
Words=[]
Common_Words =[[]] * 3

for item in BUC:
    for word in CER:
        if BUC[item]['Word']== CER[word]['Word']:
            Words.append(BUC[item]['Word'])
            Words.append(BUC[item]['Levels'])
            Words.append(CER[word]['Levels'])
    if Words!=[]:
        Common_Words.append(Words)
        Words=[]
        
    continue

Common_Words = [x for x in Common_Words if x != []]
print (Common_Words)
print (len(Common_Words))
### Saving the list
df = pd.DataFrame(Common_Words)
df.to_excel('filename.xlsx')
```
